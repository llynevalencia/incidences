<h2>Data Preparation</h2>

<h3>Try Except</h3>
<p>Catching some errors</p>

```python
col_int = ['Codi_districte', 'Codi_barri', 'NK_Any', 'Mes_any', 'Numero_incidents_GUB']

for col_i in col_int:
    try:
        df19[col_i] = df19[col_i].astype('int64')
    except ValueError as e:
        print(f"Error '{col_i}' in df19: {e}")
```

```python
df19['Mes_any'] = pd.to_numeric(df19['Mes_any'], errors='coerce').fillna(0).astype('int64')
df19['Codi_districte'] = df19['Codi_districte'].fillna(0).astype('int64')
df19['NK_Any'] = df19['NK_Any'].fillna(0).astype('int64')
df19['Numero_incidents_GUB'] = df19['Numero_incidents_GUB'].fillna(0).astype('int64')

print(df19.dtypes)
```

<h4>Unique Values</h4>
<p>Getting the unique values so that afterwards we can make the translation from catalan to english.</p>

```python
column_uq_value = ['incident_type', 'hood', 'district', 'month']

for col in column_uq_value:
    uq_value = df[col].unique()
    df_uq_value = pd.DataFrame(uq_value, columns=[col])
    output_file = f'output_data/uq_{col}.csv'
    df_uq_value.index += 1
    df_uq_value.to_csv(output_file, index=False, header=False)
```

```python
import json

with open('catalan.json', 'r') as file:
    translations = json.load(file)

for column in df.columns:
    if column in translations:
        df[column].replace(translations[column], inplace = True)
```

<h4>Top 5 Incidents by Year</h4>

<h4>Most Ocured incident by year and percentage against the rest of incidents ocured on the same year</h4>

```python
incident_count_by_year_type = df.groupby(['year', 'incident_type'])['incident_amount'].sum()

most_occurred_by_year = incident_count_by_year_type.groupby('year').idxmax()

percentage = (incident_count_by_year_type[most_occurred_by_year.values] / incident_count_by_year_type.groupby('year').sum()) * 100

result_df = pd.concat([incident_count_by_year_type[most_occurred_by_year.values], percentage], axis=1)

result_df.columns = ['Amount', '%']

result_df
```

```python

import matplotlib.pyplot as plt

incident_counts_by_year_type = df.groupby(['year', 'incident_type'])['incident_amount'].sum().reset_index()

top_incidents = incident_counts_by_year_type.groupby('incident_type')['incident_amount'].sum().nlargest(5).index

top_incidents_df = incident_counts_by_year_type[incident_counts_by_year_type['incident_type'].isin(top_incidents)]

pivot_df = top_incidents_df.pivot(index='year', columns='incident_type', values='incident_amount')

pivot_df.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Top 5 Incident Types by Year')
plt.xlabel('Year')
plt.ylabel('Incident Amount')
plt.legend(title='Incident Type')
plt.show()

```

<h4>Getting coordenates for Map in Tableau</h4>

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="my_geocoder")

uq_districts = df['district'].unique()

country = 'Spain'
state = 'Catalonia'
city = 'Barcelona'

coordinates = []

for district in uq_districts:
    location = geolocator.geocode(district + ", Barcelona, Spain")

    if location:
        latitude = location.latitude
        longitude = location.longitude
    else:
        latitude = None
        longitude = None

    coordinates.append({'country': country,'state': state,'city': city,'district': district, 'latitude': latitude, 'longitude': longitude})

coordinates_district = pd.DataFrame(coordinates)
coordinates_district.to_csv('output_data/coordinates_district.csv', index = False)
coordinates_district
```
